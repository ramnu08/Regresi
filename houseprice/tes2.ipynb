```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle 
```

```python
dataset = pd.read_csv('house_price_regression_dataset.csv')
x=dataset.drop(columns=['House_Price']).values
y=dataset['House_Price'].values
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
x_train
```

```python
x_train
```

```python
regressor.fit(x_train, y_train)
fw = open('regresi1.pkl' , 'wb')
md = pickle.dump(regressor , fw)
```

```python
y_predict = regressor.predict(x_test)
```

```python
x_test
```

```python
y_predict
```

```python
y_test
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

### Membuat Data Baru

```python
#Square_Footage, Num_Bedrooms, Num_Bathrooms, Year_Built, Lot_Size, Garage_Size, Neighborhood_Quality, House_Price
databr=dataset.drop(columns=['House_Price']).sample(10)
```

```python
print(databr.shape)
```

```python
databr
```

```python
opf = open('regresi1.pkl' , 'rb')
regres = pickle.load(opf)
y_pred = regres.predict(databr)

print(y_pred)
```

### Simpan data

```python

# Nama kolom yang Anda inginkan
columns = ['Square_Footage', 'Num_Bedrooms', 'Num_Bathrooms', 'Year_Built', 
           'Lot_Size', 'Garage_Size', 'Neighborhood_Quality']

# Simpan databr ke dalam file CSV
databr_df = pd.DataFrame(databr, columns=columns)
databr_df['Prediksi'] = y_pred  # Menambahkan hasil prediksi sebagai kolom baru

# Simpan ke file CSV
databr_df.to_csv('databr_predictions.csv', index=False)

print("Data berhasil disimpan ke dalam 'databr_with_predictions.csv'.")
```
